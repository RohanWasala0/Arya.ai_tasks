```python
from rembg import new_session

def load_model(name:str):
    _Model_PATH = 'model/u2net.onnx'
    _Model_PATH = _Model_PATH.split("/")[-1]
    _Model_PATH = _Model_PATH.split(".")[0]
    print(_Model_PATH)
    if name == _Model_PATH:
        session = new_session(model_name=_Model_PATH)
    else:
        raise RuntimeError(
            f"Model {name} not found"
        )

    return session
```

```python
SESSION = load_model('u2net')
```

```python
from rembg import remove
from PIL import Image
from io import BytesIO
import base64

with open('images.jpeg', 'rb') as f:
    doc_base64 = base64.b64encode(f.read())

decoded_data = BytesIO(base64.b64decode(doc_base64))
image = Image.open(decoded_data).convert("RGBA")

output_image = remove(image, session=SESSION).convert("RGBA")

# Save to PNG file
output_image.save("result.png", format="PNG")

# Optional: also return base64
buffered = BytesIO()
output_image.save(buffered, format="PNG")
base64_output = base64.b64encode(buffered.getvalue()).decode()
```
